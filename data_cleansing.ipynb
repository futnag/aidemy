```python
import numpy as np
import pandas as pd
np.random.seed(0)
```

```python
sample = pd.DataFrame(np.random.rand(10, 4))
sample
```

```python
sample.iloc[1, 0] = np.nan
sample.iloc[2, 2] = np.nan
sample.iloc[5:, 3] = np.nan
sample
```

```python
リストワイズ消去
データ欠損のある行をまるごと消去する
```

```python
sample.dropna()
```

```python
ペアワイズ消去
欠損の少ないデータを用いる
```

```python
sample[[0, 1, 2]].dropna()
```

```python
欠損値に代替データを代入する
```

```python
# 0で埋める
sample.fillna(0)
```

```python
# 前の値で埋める
sample.fillna(method="ffill")
```

```python
平均値代入法
欠損値をその列の平均値で埋める
```

```python
sample.fillna(sample.mean())
```

```python
マッピング
共通のキーとなるデータに対して、テーブルからそのキーに対応するデータを持ってくる処理
```

```python
a1 = {"ID": ["100", "101", "102", "103", "104", "106", "108", "110", "111", "113"]
        ,"city": ["Tokyo", "Osaka", "Kyoto", "Hokkaido", "Tokyo", "Tokyo", "Osaka", "Kyoto", "Hokkaido", "Tokyo"]
        ,"birth_year" :[1990, 1989, 1992, 1997, 1982, 1991, 1988, 1990, 1995, 1981]
        ,"name" :["Hiroshi", "Akiko", "Yuki", "Satoru", "Steeve", "Mituru", "Aoi", "Tarou", "Suguru", "Mitsuo"]}
af1 = pd.DataFrame(a1)
af1
```

```python
city_map ={"Tokyo":"Kanto"
          ,"Hokkaido":"Hokkaido"
          ,"Osaka":"Kansai"
          ,"Kyoto":"Kansai"}
city_map
```

```python
af1["region"] = af1["city"].map(city_map)
af1
```

```python
ビン分割
ビン分割とは、ある離散的な範囲にデータを分割して集計したい場合に用いる便利な機能です。
```

```python
birth_year_bins = [1980, 1985, 1990, 1995, 2000]

birth_year_cut = pd.cut(af1.birth_year, birth_year_bins)
birth_year_cut
```

```python
# ビンの数を集計
pd.value_counts(birth_year_cut)
```

```python
# それぞれのビンに名前をつける
group_names = ["first1980", "second1980", "first1990", "second1990"]
birth_year_cut = pd.cut(af1.birth_year,birth_year_bins,labels = group_names)
pd.value_counts(birth_year_cut)
```

```python
# あらかじめ分割数を指定して分割することも可能です。
# これを用いればほぼ同じサイズのビンを作成することができます。
# cutの第2引数に分割数を渡します。
pd.cut(af1.birth_year,2)
```
